Starting with prediction of attrition probability:

As the we have a categorical variable as the dependent variable the algorithms used will be:
   a: Logistic Regression
   b: Random Forest
   c: Support Vector Machine Classifiers
   
As the data set has 15,000 rows a test train validation approach will be used at 60:20:20  splits.
Ensemble of these three will be used to predict default. Prediction will be based on majority vote (2 out of 3).

5 fold Cross Validation will be used to tune hyper parameters for the models.

ROC curves will be used to measure model accuracy. Cross-Tab will also be checked to understand true/false positives and negtives.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
```

Labeling categorical data and one hot encoding

```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
```

```python
#Labelling Salary column
PreProsData=data.copy()
PreProsData["salary"]=PreProsData["salary"].map(lambda x: 0 if x=="low"  else x)
PreProsData["salary"]=PreProsData["salary"].map(lambda x: 1 if x=="medium"  else x)
PreProsData["salary"]=PreProsData["salary"].map(lambda x: 2 if x=="high"  else x)

"""
Low is tagged as o
Medium is tagged as 1
High is tagged as 2
"""
```

```python
sns.countplot(PreProsData["salary"])
```

```python
PreProsData["sales"].unique()
```

One Hot Coding the departments

```python
PreProsData["OH_Sales"]=PreProsData["sales"].map(lambda x: 1 if x=="sales"  else 0)
PreProsData["OH_accounting"]=PreProsData["sales"].map(lambda x: 1 if x=="accounting"  else 0)
PreProsData["OH_hr"]=PreProsData["sales"].map(lambda x: 1 if x=="hr"  else 0)
PreProsData["OH_technical"]=PreProsData["sales"].map(lambda x: 1 if x=="technical"  else 0)
PreProsData["OH_support"]=PreProsData["sales"].map(lambda x: 1 if x=="support"  else 0)
PreProsData["OH_management"]=PreProsData["sales"].map(lambda x: 1 if x=="management"  else 0)
PreProsData["OH_product_mng"]=PreProsData["sales"].map(lambda x: 1 if x=="product_mng"  else 0)
PreProsData["OH_marketing"]=PreProsData["sales"].map(lambda x: 1 if x=="marketing"  else 0)
PreProsData["OH_RandD"]=PreProsData["sales"].map(lambda x: 1 if x=="RandD"  else 0)
PreProsData["OH_IT"]=PreProsData["sales"].map(lambda x: 1 if x=="IT"  else 0)
```

Forming dataset for machine learning model

```python
#Splitting into Train_Test and Validation
PreProsData = PreProsData[["satisfaction_level","last_evaluation","number_project","average_montly_hours",
                          "time_spend_company","Work_accident","promotion_last_5years","salary",
                          "OH_Sales","OH_accounting","OH_hr","OH_technical","OH_support","OH_management",
                          "OH_product_mng","OH_marketing","OH_RandD","OH_IT","left","sales"]]
```

```python

X = PreProsData.iloc[:, 0:17].values
y = PreProsData.iloc[:, 18].values

# Splitting the dataset into the Training_Test and Validation set
X_train_test, X_Validation, y_train_test, y_Validation = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_train_test.shape
```

```python
X_Validation.shape
```

```python
# Splitting the dataset into the Training and Test set
X_train, X_Test, y_train, y_test = train_test_split(X_train_test, y_train_test, test_size = 0.20, random_state = 0)
```

```python
X_train.shape
```

```python
X_Test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Fitting Random Forest to training data using 3 fold cross validation

RFclassifier = RandomForestClassifier(random_state = 0)

RF_param = { 
    'n_estimators': [300, 500, 700],
    'max_features': ["sqrt", "log2"]
}

RF_Grid_Search = GridSearchCV(estimator=RFclassifier, param_grid=RF_param, cv= 3)

RF_Grid_Search.fit(X_train, y_train)
```

```python
#Identifying the best parameters for Random Forest
print(RF_Grid_Search.best_params_)
```

```python
#Building Random forest model with the identified best parameters
RF_final_classifier = RandomForestClassifier(n_estimators = 500, max_features = 'sqrt', random_state = 0)
RF_final_classifier.fit(X_train, y_train)

# Predicting the Test set results
y_test_pred = RF_final_classifier.predict(X_Test)

# Checking Accuracy on Test data
RF_roc_auc = roc_auc_score(y_test, y_test_pred)
print ("Test AUC for Random Forest = %2.2f" % RF_roc_auc)
print(classification_report(y_test, y_test_pred))
```

```python
#ROC Curve for Random Forest

fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)

plt.figure()
plt.plot(fpr, tpr, label='ROC Cure (area = %0.2f)' % RF_roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Graph')
plt.legend(loc="lower right")
plt.show()
```

```python
#Fitting SVM to training data using 3 fold cross validation

SVMclassifier = SVC(random_state = 0)

SVM_parameters = [{'C': [1, 10], 'kernel': ['linear']},
              {'C': [1, 10], 'kernel': ['rbf'], 'gamma': [0.1, 0.2, 0.3]}]

SVM_grid_search = GridSearchCV(estimator=SVMclassifier, param_grid=SVM_parameters, cv= 3)

SVM_grid_search.fit(X_train, y_train)
```

```python
print(SVM_grid_search.best_params_)
```

```python
#Building SVM model with the identified best parameters

SVM_final_classifier = SVC(C=10, kernel="linear", gamma=0.1)
SVM_final_classifier.fit(X_train, y_train)

# Predicting the Test set results
y_test_pred = SVM_final_classifier.predict(X_Test)

# Checking Accuracy on Test data
SVM_roc_auc = roc_auc_score(y_test, y_test_pred)
print ("Test AUC for Support Vector Machine = %2.2f" % SVM_roc_auc)
print(classification_report(y_test, y_test_pred))
```

```python
#ROC Curve for Support Vector Machine

fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)

plt.figure()
plt.plot(fpr, tpr, label='ROC Cure (area = %0.2f)' % SVM_roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Graph')
plt.legend(loc="lower right")
plt.show()
```
